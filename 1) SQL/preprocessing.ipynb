```python
import pandas as pd
```

```python
DATA_DIR = "./data/" # путь до папки с данными
```

```python
data = pd.read_csv(f"{DATA_DIR}Таблица_SQL.csv", sep=";", encoding="windows-1251") # считываем данные, разделённые с помощью sep и имеющие кодировку encoding
data
```

```python
print(data.columns[4:])  # удаляемые столбцы

data.drop(columns=data.columns[4:], inplace=True) # удаляем столбцы, не несущие никакой информации
data
```

```python
data.to_csv(f"{DATA_DIR}table_refactored.csv", sep=",", encoding="utf-8", index=False) # сохраняем данные без лишних столбцов, в кодировке UTF-8 и со стандартным разделителем (запятой) (и без автоматически созданного столбца индексов)
```
